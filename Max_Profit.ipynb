```python
def max_profit(n):
    # Definitions
    # index 0:Theatre(T)
    # index 1:Pub(P)
    # index 2:Commertial park (C)
    build_time=[5,4,10]
    prize=[1500,1000,2000]
    labels=['T','P','C']

    # f=max earing from time t to n
    f=[0]*(n+1)
    # Which building to start at time t
    choice=[-1]*(n+1)


    # finding max earning form t to n
    for t in range(n-1,-1,-1):
        best=0
        best_choice=-1
        for i in range(3):
            finish= t + build_time[i]
            if finish <= n:
                earn_now = prize[i] *max(0,n-finish)
                earn_future=f[finish]
                total = earn_now + earn_future

                if total > best:
                    best=total
                    best_choice= i

        f[t] = best
        choice[t]=best_choice


    counts=[0,0,0]
    t=0
    while t<n and choice[t] != -1:
        i= choice[t]
        counts[i] += 1
        t += build_time[i]

    result={
        'max_earning': f[0],
        'T':counts[0],
        'P':counts[1],
        'C':counts[2]
    }
    return f[0],counts
```

```python
# input
n=int(input("Enter The Time Units: "))

profit,counts= max_profit(n)

print("\n===RESULT===")
print(f"Earning: {profit}")
print(f"T:{counts[0]} P:{counts[1]} C:{counts[2]}")
```
